```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
```

```python
iris.head()
```

```python
iris['class'] = iris['class'].astype('category')
```

```python
sns.scatterplot(x=iris.sepal_length, y=iris.sepal_width,hue=iris['class'] )
```

```python
sns.scatterplot(x=iris.petal_length, y=iris.petal_width,hue=iris['class'] )
```

```python
sns.scatterplot(x=iris.sepal_length, y=iris.petal_length,hue=iris['class'] )
```

```python
sns.boxplot(iris)
```
